### 4.1 Calculo de VWAP (Volume-Weighted Average Price)

Esta métrica es muy utilizada y es de las más populares al rededor del mundo, el calculo consiste en multiplicar los valores bid vs los volumnes bid, despúes sumar estos valores, hacer los mismo con los precios y volumendes del ask, después de esto tendremos los dos precios ponderados del bid y el ask por lo que los sumamos y finalmente este número lo dividimos entre la suma total de los volumenes del bid y el ask

### 4.2 Calculo de Order Book Imbalance

Esta métrica es útil para conocer cual es el balance de las ordenes en el libro, es decir que porcentaje de las ordenes pertenecen a ordenes de venta y cuales a las de compra y se calcula dividiendo el Bid Volume / Ask Volume, lo que nos dara como resultado el porcentaje las ordenes Bid por lo que el restante será lo que pertenezca a las ordenes de Ask
